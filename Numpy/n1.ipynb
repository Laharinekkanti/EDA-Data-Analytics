Arrays works faster than the lists,so numpy uses array

```python
l=[10,20,30,40,50]
type(l)
a1=np.array(l)
a1
```

```python
l1=[10,2,3.5,7,'np','6j+c']
a2=np.array(l1)
type(a2)
print(a2)
```

Arrays accepts only homogeneous data

```python
#when we try to add 2 lists it will concatenate the elements from l1 to l2
l1=[10,20,30,40,50]
l2=[1,2,3,4,5]
print(l1+l2)
```

```python
#when we try to add array elements,it will sum up the elements present in same index value
arr1=np.array(l1)
arr2=np.array(l2)
print(arr1+arr2)
```
