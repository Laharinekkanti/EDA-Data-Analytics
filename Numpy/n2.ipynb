### linspace-->divides the numbers into 50(default) parts equally

linspace(start,step,num,,endpoints)
- `num`=number how many parts we have to divide
- `retstep`=return step values

```python
np.linspace(0,10)
np.linspace(0,10,num=5) #array([ 0. ,  2.5,  5. ,  7.5, 10. ])
np.linspace(0,10,num=5,retstep=True) #(array([ 0. ,  2.5,  5. ,  7.5, 10. ]), np.float64(2.5))
np.linspace(0,10,num=5,retstep=True,endpoint=False) #(array([0., 2., 4., 6., 8.]), np.float64(2.0))
```

## Array Creation Methods

### Zero function

- It creates matrix containing only zero
- syntax: np.zeroes(shape,dtype=float,order='C')
- return type: ndarray

```python
np.zeros(2)
np .zeros((2,3))
np.zeros((2,3,4,)) #2-blocks,3-rows,4-columns
```

### Ones function

- It creates matrix containing only one
- syntax: np.zeroes(shape,dtype=float,order='C')

```python
np.ones(2)  #array([1., 1.])
np.ones((2,3)) ##2-rows,3-columns
np.ones((2,3,5)) #2-blocks,3-rows,5-columns
```

### Eye function

- generates square matrix in which diagonal contains 1 rest all will be zero(0).

```python
np.eye(3)
np.eye(4,4)
```

### Full function
- fill the matrix with the same element which we have given
- syntax: np.full(size,element)

```python
np.full(2,5) #array([5, 5])
np.full((2,4),100) #2-rows,4-columns,element-100
```

### Random Module
- used to generate random numbers

```python
import numpy.random as rd
```

```python
#gives numbers between 0 to 2 , 2 is excluded
rd.randint(2)
```

```python
rd.randint(10,100) #generates numbers between 10 to 100
rd.randint(10,100,(3)) #generates matrix of 3 columns
rd.randint(10,100,(3,2)) #generates matrix with random numbers of 3-rows and 2-columns
```

### Seed method
- seed value used to fix the values instead of generating random numbers everytime we execute

```python
rd.seed(2)
rd.randint(10,100,(3,2))

print(rd.randint(10,100,(3,2)))
```

### rand() method
- generates random numbers between 0-1 (float values)
- generates only positive values
- syntax: rd.rand(size)

```python
rd.rand(3)
rd.rand(2,3) #2-rows, 3-columns
```

### randn() method
- generates random numbers between 0-1 (float values)
- generates both positive and negative values
- generated random numbers mean=0, standard deviation=1
- syntax: rd.randn(size)

```python
r=rd.randn(2)
print(r)

print(np.mean(r))
print(np.std(r))
```

```python
r1=rd.randn(2,3)
print(r1)

print(np.mean(r1))
print(np.std(r1))
```

### choice() method
-used to choose the elements which we want

```python
l=[10,20,30,40,50]
rd.choice(l) #np.int64(40)
rd.choice(l,3,replace=False) # array([40, 30, 50])   #replace-->used to avoid duplicate elements
rd.choice(l,(2,3))
```

```python
l1=['purple','red','yellow','maroon','pink','green','blue','white']
rd.choice(l1) #np.str_('yellow')
rd.choice(l1,3,replace=False) #array(['maroon', 'purple', 'white'], dtype='<U6')
rd.choice(l1,3,replace=False,p=[0.17,0.1,0.12,0,0,0.11,0.5,0]) #sum of the values should be equal to=1
```
