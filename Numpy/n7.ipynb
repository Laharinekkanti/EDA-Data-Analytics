``AGGREGATE/STATISTICAL METHODS``

```python
import numpy as np
```

```python
arr1=np.array([10,20,30])
print(arr1)
```

`power()`:used to square all the elements present in the array 

```python
print(np.power(arr1,2))
```

`sum()`

```python
print(np.sum(arr1))
```

`standard deviation`

```python
print(np.std(arr1))
```

`mean()`

```python
print(np.mean(arr1))
```

`variance()`:used to check how far the value is from mean value

```python
print(np.var(arr1))
```

```python
arr2=np.array([1,0,2])
print(arr2)
```

`exp()`:e=2.734

```python
print(np.exp(arr2))
```

`log()`:gives the value of particular power

```python
arr3=np.array([2.71828183 ,1.         ,7.3890561 ])
print(arr3)
print(np.log(arr3))
```

```python
arr4=np.array([10,2,15,5,9,20])
print(arr4)
```

`isin()`: works same as membership operator(in)

```python
print(np.isin(20,arr4))
print()
print(np.isin([2,15,8,20],arr4))
```

`sort()`

```python
print(np.sort(arr4))
```

`median()`:gives mid value

```python
print(np.median(arr4))
```

`percentile()`

```python
print(np.percentile(arr4,25))
print(np.percentile(arr4,50))
print(np.percentile(arr4,75))
print(np.percentile(arr4,100))
```

`where()`


```python
print(arr4[np.where(arr4<11)])
```
