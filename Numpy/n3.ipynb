```python
import numpy as np
```

```python
lst1=[10,20,30,40,50]
lst1[1] #20
arr1=np.array(lst1)
arr1
```

```python
lst2=[[10,20,30,40],[50,60,70,80],[90,18,21,11]]
lst2[1][1] #60
# '''
# lst2[1,1] 
# TypeError: list indices must be integers or slices, not tuple
# '''

arr2=np.array(lst2)
arr2
```

```python
# array supports multi-indexing
print(arr1[2]) 
print(arr2[1,1]) 
print(arr2[1][1]) 
print(arr2[1][3])
```

### Slicing()

```python
#using list
print(lst1[::-1])
```

```python
print(lst2[::-1])
print(lst2[::-1][2])
```

```python
#using array
print(arr2[::-1])
print()
print(arr2[::,::-1])
print()
print(arr2[::-1,2])
print()
print(arr2[::-1,::-1])
print()
print(arr2[::-1][2])
print()
print(arr2[::2,::2])
print()
print(arr2[:1:,::2])
```

```python
lst3=[[[10,20],[30,40]],[[1,2],[50,60]]]
arr3=np.array(lst3)
arr3
```

```python
print(arr3[1,0,1])
print()
print(arr3[::,::,::])
print()
print(arr3[::-1,:1:,::2])
```
