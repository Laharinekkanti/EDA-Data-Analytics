```python
import numpy as np
```

```python
arr1=np.arange(10,96,5)
print(arr1)
print(arr1.shape)
```

`reshape() `
- used to convert the one shape of arary to another shape of array
- to change the shape of array we have to give shape=(mutiples of shape of array)-->eg=18(multiples of 18)
- (-1)--> converts any shape of the array to 1D array

```python
print(np.reshape(arr1,shape=(2,9)))
print()
print(np.reshape(arr1,shape=(3,6)))
```

```python
arr2=np.reshape(arr1,(3,2,3))
print(arr2)
print(np.reshape(arr2,(-1))) 
```

```python
arr3=np.array([[10,20,30,40,50,78],[2,4,5,60,64,28],[11,9,37,92,55,68]])
print(arr3)
print(arr3.shape)
print(arr3.ndim,arr3.size,sep='\n')
print()
```

```python
print(np.reshape(arr3,(3,2,1,3)))
print()
print(np.reshape(arr3,(3,3,2)))
```
