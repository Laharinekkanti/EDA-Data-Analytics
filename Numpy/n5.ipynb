`Broadcasting`-->applying one value for each and every element. 

Rules
- elements must be equal(same size)-->ele==ele
- elements should be equal to 1 or empty--> ele==1 and ele==empty

```python
import numpy as np
```

```python
l1=[10,20,30]
arr1=np.array(l1)
print(arr1)
print(arr1.shape)
c=100
```

```python
arr1 + c
```

```python
l2=[11,12,13]
arr2=np.array(l2)
print(arr2)
print(arr2.shape)
```

```python
arr1+arr2
```

```python
arr3=np.array([100,200,300,400])
print(arr3)
```

```python
'''
arr1+arr3
ValueError: operands could not be broadcast together with shapes (3,) (4,) 
'''
```

```python
arr4=np.array([[1,2,3],[5,6,7]])
print(arr4)
print(arr4.shape)
```

```python
arr1+arr4
```

```python
arr5=np.array([[[8],[5]],[[2],[1]],[[23],[36]]])
print(arr5)
print(arr5.shape)
```

```python
arr4+arr5
```
