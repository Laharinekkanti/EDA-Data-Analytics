```python
import numpy as np
```

```python
arr1=np.array([[1,2,3],[10,20,30]])
print(arr1)
```

```python
print(np.reshape(arr1,(-1)))
```

`ravel()`
- converts any shape of the array into 1D array(returns a view)
- if any changes made in new array ,it also changes in original array

```python
x=arr1.ravel()
print(x)
print()
x[0]=50
print(arr1,x,sep='\n')
```

```python
arr2=np.array([[11,22,33],[4,67,80]])
print(arr2)
```

`flatten()`
- Flattens a multi-dimensional array to a 1D array (returns a copy). 

```python
y=arr2.flatten()
print(y)
print()
y[-1]=900
print(arr2,y,sep='\n')
```
