## sort()

used to arrange the data in ascending order or descending order

```python
df=df.sort_values(by=['ticket','age'],ascending=[True,True])
```

```python
df=df.reset_index(drop=True)
```

```python
ticket_pass=df['ticket'].value_counts().reset_index(name='no_of_passengers')
ticket_pass
```

## merge()

used to join the 2 tables

syntax-->pd.merge(on=,right_on=,left_on=,how=)

on-->column_name

left_on-->cloumn_name of left table

right_on-->column_name of right table

how-->inner,outer(outerjoin),left(left outer join),right(right outer join)

```python
df=pd.merge(df,ticket_pass,on='ticket',how='inner')
```

```python
df['fare']=(df['fare']/df['no_of_passengers']).round(1)
```

```python
def comp_type(row):
    f=row['Fam_count']
    p=row['no_of_passengers']
    if p==1:
        return 'Solo'
    elif f==0 and p>1:
        return 'Friends'
    else:
        return 'Family'
```

## apply()--> used to customise the data row by row and used for complex function

```python
# comp_type--> user_defined function
df['Companion type']=df.apply(comp_type,axis=1)
```

## cut()-->used only for numerical data 

## it is a pandas built-in method,also used for continuous values

```python
# def age_cat(row):
#     if age>0 and age<=2:
#         return 'infant'
#     elif age>2 and age<=16:
#         return 'child'
#     elif age>16 and age<=25:
#         return 'teen'
#     elif age>25 and age<=59:
#         return 'adult'
#     else:
#         return 'Senior'

df['Age category']=pd.cut(df['age'],bins=[0,2,16,25,59,100],labels=['Infant','Child','Teen','Adult','Senior'])
```

```python
Royal=['the Countess','Jonkheer','Don','Dona','Master','Sir']
Common=['Miss','Mr','Mrs','Lady','Ms','Mme','Mlle']
Military=['Major','Col','Capt']
Other_professional=['Dr','Rev']
```

```python
def social_status(row):
    c=row['Title']
    if c in Royal:
        return 'Royal'
    elif c in Common:
        return 'common'
    elif c in Military:
        return 'Military'
    else:
        return 'Other_professional'
df['Social status']=df.apply(social_status,axis=1)
```

### count()

```python
plt.figure(figsize=(3,4)) # width, Height
sns.countplot(data=df,x='Gender',color='violet')
plt.title('No of passengers gender wise')
plt.yticks(range(0,1000,100)) #to change the ranges
plt.ylabel('No of passengers')
plt.show()  #only shows figure without unwanted text
```

```python
plt.figure(figsize=(4,5))
cnt=sns.countplot(data=df,x='pclass',color='orange',order=[3,1,2])#order--> used to arrange the bar in ascending or descending order
cnt.bar_label(cnt.containers[0])#bar_labels--> to give the count of each caterogy
plt.title('No of passengers in each class')
plt.yticks(range(0,900,100))
plt.ylabel('No of passengers')
plt.show() 
```

**Observation**

pclass 3 has more no of passengers compared to other classes

```python
plt.figure(figsize=(5,4))
c=sns.countplot(data=df,x='Age category',hue='Gender',palette=['cyan','crimson'])
#c.bar_label(c.containers[1])
for i in c.containers:
    c.bar_label(i)
plt.title('Distribution of passengers according to Age category')
plt.yticks(range(0,1000,100))
plt.ylabel('No of passengers')
plt.show() 
```

**Observation**

There are more no males compared to females based on Age category

### BAR PLOT

used for categorical data as well as numerical data

it can be bivariant or univariant

2 columns-->used to check the relationship b/w the columns

1 column-->used to check the behaviour of that column

used to check the aggregate data

Barplot always gives the mean value

```python
df.head()
```

```python
# categorical should always be used in x axis
#to remove the errorbar we have to use parameter called errorbar
#estimator-->used for passing aggregate functions
#fmt--> used for formatting the numbers
plt.figure(figsize=(3,4))
c=sns.barplot(data=df,x='Gender',y='fare',errorbar=None,estimator='mean')
c.bar_label(c.containers[0],fmt='%.1f')
plt.title('Avg fare by Gender wise')
plt.yticks(range(0,20,2))
plt.show()
```

**Observation**

Female passengers has higher avg_fare than compared to male

```python
plt.figure(figsize=(3,4))
c=sns.barplot(data=df,x='Gender',y='fare',errorbar=None,estimator='sum')
c.bar_label(c.containers[0],fmt='%.1f')
plt.title('Avg fare by Gender wise')
plt.yticks(range(0,14000,2000))
plt.show()
```

**Observation**

Males total_fare is higher,compared to female passengers

```python
#hue-->used to make the subgroups...
plt.figure(figsize=(5,4))
c=sns.barplot(data=df,x='Age category',y='fare',errorbar=None,estimator='mean',hue='Gender')
for i in c.containers:
    c.bar_label(i,fmt='%.1f')
plt.title('Avg fare of passengers according to Age category')
plt.yticks(range(0,40,5))
plt.legend(bbox_to_anchor=(1,1),loc='best',ncol=1,title='GENDER')# used to place the legend in best position
plt.show()
#ncol=2--->used to place the columns beside one another
#bbox_anchor_tag-->used for adjusting the legend box(left,right...)
```

**Observation**

Avg fare increase with the age, among all the categories females pays consistently higher avg_fare compared to males

#### SUB PLOT

plotting multiple graphs in a single page

```python
g=df.loc[df['Companion type']=='Family']
gf=g['Gender'].value_counts()
gn=g['Age category'].value_counts()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.suptitle("Companion typewise categorization")
axes[0].pie(x=gf.values,labels=gf.index,autopct='%.1f%%')
axes[0].set_title('Gender wise categorization')
axes[1].pie(x=gn.values,labels=gn.index,autopct='%.1f%%')
axes[1].set_title('Age category wise categorization')
plt.tight_layout()# used to adjust the space....
plt.show()
```

**Observation**

In a family most of the passengers are females and most of the passengers are adults..
