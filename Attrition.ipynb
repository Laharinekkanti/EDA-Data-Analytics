## Data Cleaning

-->checking for null values

-->checking duplicates

```python
df.duplicated(subset='EmployeeID').any()
```

```python
df=df.drop_duplicates(subset='EmployeeID',keep='last')
```

```python
df.shape
```

```python
col=['recorddate_key','birthdate_key','orighiredate_key','terminationdate_key']
for i in col:
    df[i]=pd.to_datetime(df[i])
```

```python
df.info()
```

```python
df=df.reset_index(drop=True)
```

```python
df=df.drop(columns=['store_name','gender_short'])
```

```python
df.columns
```

```python
col_list=['ID','record date','DOB','hire date','term date','age','experience','city','dname','job','gender','term reason','term type','year','status','business unit']
df.columns=col_list
```

```python
df
```

```python
df['status']=df['status'].str.title()
df['business unit']=df['business unit'].str.title()
```

```python
# to find min age and max age
df[['age','experience']].agg(['min','max'])
```

```python
#cut()--> used to continuous range of numerical values.. and it is faster

df['age category']=pd.cut(df['age'],bins=[18,25,56,65],labels=['Teen','Adult','Senior'])
```

```python
df['service category']=pd.cut(df['experience'],bins=[-1,1,5,26],labels=['Fresher','Senior','Experts'])
```

```python
df['dname'].unique()
```

```python
df['job'].unique()
```

```python
def job_clean(row):
    j=row['job']
    d=row['dname']
    if j!='Auditor' and j!='Dairy Person':
        if d in j:
            return j.replace(d,'').strip().strip(',').strip(', s')
        else:
            return j.removeprefix('Exec Assistant').strip(', ')
    else:
        return j
```

```python
df['job']=df.apply(job_clean,axis=1)
```

```python
df['job'].unique()
```

```python
df['job']=df['job'].replace('CHief Information Officer','Chief Information Officer')
```

```python
h_df=df['hire date'].dt.year.value_counts().sort_values().reset_index()
h_df.columns=['year','hiring']
h_df
```

```python
t_df=df.loc[df['status']=='Terminated','term date'].dt.year.value_counts().sort_values().reset_index()
t_df.columns=['year','termination']
t_df
```

## Merging of 2 tables

```python
df1=pd.merge(h_df,t_df,on='year',how='outer').fillna(0).astype(int)
```

```python
df1['retention1']=df1['hiring']-df1['termination']
df1
```

#### Cummulative Sum

```python
df1['retention']=df1['retention1'].cumsum()
```

```python
df1=df1.drop(columns='retention1')
```

## - Kernal Density estimator
- continuous plot
- used for continuous range of values
- only used for numerical data
- it gives the distribution of data
- it gives the result in form of density 
- here we can predict exact values using the curves
- it doesn't count the values,it gives the density

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df,x='experience')
plt.xticks(range(0,30))
plt.grid()
```

***Observation***
- Based on the kernel density estimator graph we can conclude that there are more employees with 13 years experience.

### Boxplot(univariant,bivariant)

- used to check the distribution of data
- used to identify outliers and to fill the null values
- consists of outliers(datapoints far away from the majority of data) and distribution(spread of data)
- consists of quartile region(Q1,Q2,Q3)
- Inter quartile range=Q3-Q1
- outliers-->beyond lower and upper bound
- if there is any outliers in symmetric plot-->fill it with mean values
- for right-skewed-->fill with median values
- for left-skewed-->fill with median values

```python
l=[220,225,272,215,1,10,990,900,203,255,210]
s=pd.Series(l)
sns.boxplot(x=s)
```
