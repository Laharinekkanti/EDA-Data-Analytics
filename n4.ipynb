#### operators in numpy

```python
import numpy as np
```

```python
l1=[1,2,3]
arr1=np.array(l1)
arr1
```

```python
l2=[6,7,8]
arr2=np.array(l2)
arr2
```

##### Arithematic Operations-->(+, -, *, **, %, /, //)

```python
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1**arr2)
print(arr1/arr2)
print(arr1//arr2)
print(arr1%arr2)
```

##### Comparision/Relational operators-->(==, !=, <, >, <=, >=)

```python
print(arr1==arr2)
print(arr1!=arr2)
print(arr1<arr2)
print(arr1>arr2)
print(arr1<=arr2)
print(arr1>=arr2)
```

- `all()`: returns True when all conditions are True
- `any()`: returns True if any one condition is True

```python
print((arr1>5).all() and (arr1<10).all())
print((arr2>5).all() and (arr2<10).all())
```

```python
print((arr1>5).any() and (arr1<10).any())
print((arr2>5).any() and (arr2<10).any())
```
